Use Matplotlib to create bar charts that visualize the conclusions you made with groupby and query.

```python
# Import necessary packages and load `winequality_edited.csv`
import matplotlib.pyplot as plt
% matplotlib inline
import pandas as pd

df = pd.read_csv('winequality_edited.csv')
```

```python
df.columns
```

### #1: Do wines with higher alcoholic content receive better ratings?
Create a bar chart with one bar for low alcohol and one bar for high alcohol wine samples. This first one is filled out for you.

```python
# Use query to select each group and get its mean quality
median = df['alcohol'].median()
low = df.query('alcohol < {}'.format(median))
high = df.query('alcohol >= {}'.format(median))

mean_quality_low = low['quality'].mean()
mean_quality_high = high['quality'].mean()
```

```python
# Create a bar chart with proper labels
locations = [1, 2]
heights = [mean_quality_low, mean_quality_high]
labels = ['Low', 'High']
plt.bar(locations, heights, tick_label=labels)
plt.title('Average Quality Ratings by Alcohol Content')
plt.xlabel('Alcohol Content')
plt.ylabel('Average Quality Rating');
```

### #2: Do sweeter wines receive higher ratings?
Create a bar chart with one bar for low residual sugar and one bar for high residual sugar wine samples.

```python
# Use query to select each group and get its mean quality
#finding median of residual sugar
median_sugar = df['residual sugar'].median()
df_sweet = df[df['residual sugar'] < median_sugar]
df_sweeter = df[df['residual sugar'] > median_sugar]

mean_quality_sweet = df_sweet['quality'].mean()
mean_quality_sweeter = df_sweeter['quality'].mean()
```

```python
# Create a bar chart with proper labels
locations = [1, 2]
heights = [mean_quality_sweet, mean_quality_sweeter]
labels = ['Sweet', 'Sweeter']
plt.bar(locations, heights, tick_label=labels)
plt.title('Average Quality Ratings by Sugar Content')
plt.xlabel('Sugar Content')
plt.ylabel('Average Quality Rating');
```

### #3: What level of acidity receives the highest average rating?
Create a bar chart with a bar for each of the four acidity levels.

```python
# Use groupby to get the mean quality for each acidity level
df.groupby('acidity_levels').mean()['quality']
```

```python
# Create a bar chart with proper labels
# Create a bar chart with proper labels
locations = [1, 2,3,4]
heights = df.groupby('acidity_levels').mean()['quality']
labels = ['High', 'low','Medium','Moderately High']
plt.bar(locations, heights, tick_label=labels)
plt.title('Average Quality Ratings by Acidity level')
plt.xlabel('Acidity Level')
plt.ylabel('Average Quality Rating');
```
